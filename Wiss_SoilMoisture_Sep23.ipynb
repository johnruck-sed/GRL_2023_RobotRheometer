```python
import numpy as np # for math
import pandas as pd # to load the spreadsheet
from matplotlib import pyplot as plt # for plotting

#load data
moisture_data = np.asarray(pd.read_csv('Data/JupyterWissData.csv'))

#moisture and distance for both Transects A & B
dist_vals = moisture_data[:,2]
reverse_dist_vals = dist_vals[::-1]
moisture_vals = moisture_data[:,4]
reverse_moisture_vals = moisture_vals[::-1]

#plotting
plt.scatter(reverse_dist_vals, moisture_vals, color = 'b')
plt.xlabel('Horizontal Distance (m)')
plt.ylabel('Moisture (V)')
plt.show()
```

```python
#separating Transect A & B data
sep23moisture = np.asarray(pd.read_csv(('Data/Wiss_Sep23.csv')))
moisture_TransectB = sep23moisture[0:38,4]
moisture_TransectA = sep23moisture[38:,4]

distance_TransectB = sep23moisture[0:38,2]
distance_TransectA = sep23moisture[38:,2]
reverse_distance_TransectB = distance_TransectB[::-1]
reverse_distance_TransectA = distance_TransectA[::-1]

plt.scatter(reverse_distance_TransectA, moisture_TransectA, color = 'r', label = 'Transect A')
plt.scatter(reverse_distance_TransectB, moisture_TransectB, color = 'b', label = 'Transect B')
plt.xlabel('x (m)') #horizontal distance downslope
plt.ylabel('moisture (V)') #moisture (voltage)
plt.legend(fontsize = 'small')
plt.show()
```

```python
#plot soil moisture against d50 for both transects A & B

#A
avgsample_d50_data = np.asarray(pd.read_csv('Data/Sample_D50_good.csv'))
ddvals = avgsample_d50_data[:,::2]

#B
B_d50 = [98.295, 104.028, 84.012, 80.032, 84.961, 89.951, 91.859, 133.402, 98.874, 117.386, 97.677, 130.772, 164.124, 119.323, 118.906, 136.906, 129.37, 128.838, 99.879,114.877,130.949,149.858,125.611,122.6,129.115,110.423,147.116,135.713,110.071,154.399,104.003,117.849,157.376,159.435,96.607,135.439,82.02,109.352]

#plot
plt.scatter(ddvals, moisture_TransectA, color = 'r', label = 'Transect A')
plt.scatter(B_d50, moisture_TransectB, color = 'b', label = 'Transect B')
plt.xlabel('soil particle d50 (\u03BCm)') #d50
plt.ylabel('moisture (V)') #moisture (voltage)
plt.legend(fontsize = 'small')
plt.show()
```

```python
#plotting soil moisture against distance downslope
#during and after rain storm???? (supplemental fig) (July 29/30)

#load csv
WissData = np.asarray(pd.read_csv('Data/All_Wiss_Data.csv'))

#pull out July data
July29_moist = WissData[52:88,4]
July30_moist = WissData[91:123,4]

#pull out downslope distance data (x)
July29_dist = WissData[52:88,2]
July30_dist = WissData[91:123,2]

#plot
plt.scatter(July29_dist, July29_moist[::-1], label = 'July 29')
plt.ylabel('moisture (V)')
plt.xlabel('x (m)')
plt.legend()

plt.scatter(July30_dist, July30_moist[::-1], label = 'July 30')
plt.legend()
```

```python
#plotting soil moisture against distance downslope during and after rain events
##(supplemental fig) (Aug 6/8)

#dry
Aug5_moist = WissData[123:135,4]
Aug5_dist = WissData[123:135,2]

#rain
Aug6_moist = WissData[135:147,4]
Aug6_dist = WissData[135:147,2]

#no rain
Aug8_moist = WissData[147:159,4]
Aug8_dist = WissData[147:159,2]

#plot
plt.scatter(Aug5_dist, Aug5_moist[::-1], label = 'August 5')
plt.scatter(Aug6_dist, Aug6_moist[::-1], label = 'August 6')
plt.scatter(Aug8_dist, Aug8_moist[::-1], label = 'August 8')
plt.ylabel('moisture (V)')
plt.xlabel('x (m)')
plt.xlim(0,200)
plt.ylim(1,2.5)
plt.legend()
```

```python
#plotting soil strength against distance downslope during and after rain event
##(supplemental fig) (Aug 6/8)

#pull strength data
Aug5_s = WissData[123:135,5]
Aug6_s = WissData[135:147,5]
Aug8_s = WissData[147:159,5]

#plot
plt.scatter(Aug5_dist, Aug5_s[::-1], label = 'August 5')
plt.scatter(Aug6_dist, Aug6_s[::-1], label = 'August 6')
plt.scatter(Aug8_dist, Aug8_s[::-1], label = 'August 8')
plt.ylabel('k')
plt.xlabel('x (m)')
plt.xlim(0,200)
plt.legend()
```
