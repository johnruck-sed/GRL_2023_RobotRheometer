```python
#Averaged Main Transect

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pylab as pl
from matplotlib import cm
import matplotlib.colorbar

df = np.asarray(pd.read_csv('Data\GrainSizeAverage.csv',header=1))
gvals_av = df[:,::2]
vvals_av = df[:,1::2]


def colorlist(color1, color2, num):
    """Generate list of num colors blending from color1 to color2"""
    result = [np.array(color1), np.array(color2)]
    while len(result) < num:
        temp = [result[0]]
        for i in range(len(result)-1):
            temp.append(np.sqrt((result[i]**2+result[i+1]**2)/2))
            temp.append(result[i+1])
        result = temp
    indices = np.linspace(0, len(result)-1, num).round().astype(int)
    return [result[i] for i in indices]

colors = colorlist((1, 0, 0), (0, 0, 1), len(gvals))

for i in range(len(gvals)-1):
    xi = gvals[i:i+1+1]
    yi = vvals[i:i+1+1]
    ci = colors[i]
    #plt.xscale("log")
    #plt.xlabel('Grain Size(μm)')
    #plt.ylabel('Average Diff. % Volume(%)')
    #plt.plot(xi, yi, color=ci, linestyle='solid', linewidth='1')


plt.plot(gvals_av,vvals_av)
plt.title('Averaged Main Transect')
plt.xscale('log')
plt.xlabel('grain size(μm)')
plt.ylabel('diff. volume (%)')
plt.text(1000, 5, 'n = 38', fontsize = 8, color = 'black')
plt.show()
```
